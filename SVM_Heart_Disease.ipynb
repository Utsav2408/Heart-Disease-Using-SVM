```
import pandas as pd
```

```
df = pd.read_csv('HD_Cleveland_Dataset.csv')
df
```

```
df['target'] = pd.get_dummies(df['target'], drop_first=True)
```

```
df
```

```
df['target'].value_counts()
```

```
df['target'].nunique()
```

```
from sklearn.model_selection import train_test_split
X = df.drop('target',axis=1)
y = df['target']
```

```
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=101)
```

```
# Importing Trating & Support Vector Classfier

from sklearn.svm import SVC
```

```
# Create Instance

svm_model = SVC()
```

```
# Training / Fitting Model
svm_model.fit(X_train,y_train)
```

```
# Prediction

predictions = svm_model.predict(X_test)
```

```
from sklearn.metrics import classification_report,confusion_matrix
```

```
print(confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))
```

```
param_grid = {'C':[0.01,0.1,1.0,10,100,1000,10000,100000],
              'gamma':[1,0.1,0.01,0.001,0.0001,0.00001,0.000001]}
```

```
# We have to import gridsearchcv from model_selection - CV - Cross Validation

from sklearn.model_selection import GridSearchCV
```

```
grid = GridSearchCV(SVC(),param_grid,verbose=2)
```

```
# Lets Try to Fit the Data Now 

grid.fit(X_train,y_train)
```

```
grid_predictions = grid.predict(X_test)
```

```
print(confusion_matrix(y_test,grid_predictions))
print(classification_report(y_test,grid_predictions))
```
